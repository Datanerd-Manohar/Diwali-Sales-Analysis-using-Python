# Diwali Sales Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\user\Downloads\Python_Diwali_Sales_Analysis-main\Python_Diwali_Sales_Analysis-main\Diwali Sales Data.csv', encoding ='unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.head()
```

```python
df[['Amount','Age','Orders']].describe()
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gen_sales = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot( x = 'Gender', y = 'Amount', data = gen_sales)
```

```python
df.info()
```

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_sales = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = age_sales)
```

```python
state_amount_sales = df.groupby('State', as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize': (25,5)})
sns.barplot(x = 'State', y = 'Amount', data = state_amount_sales)
```

```python
state_order_sales = df.groupby('State', as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

sns.set(rc={'figure.figsize': (25,5)})
sns.barplot(x = 'State', y = 'Orders', data = state_order_sales)
```

```python
df.columns
```

```python
martial_sales = sns.countplot(x = 'Marital_Status', data = df, hue = 'Gender')


sns.set(rc={'figure.figsize': (4, 8)})
for bars in martial_sales.containers:
    martial_sales.bar_label(bars)
```

```python
status_amount_sales = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc={'figure.figsize' : (4,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = status_amount_sales)
```

```python
status_orders_sales = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

sns.barplot(x = 'Marital_Status', y = 'Orders', hue = 'Gender', data = status_orders_sales)
```

```python
sector_sales = df.groupby('Occupation', as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

sns.set(rc={'figure.figsize': (25,5)})
sns.barplot(x = 'Occupation', y = 'Orders', data = sector_sales)
```

```python
count = sns.countplot(x = 'Occupation', data = df)

for bars in count.containers:
    count.bar_label(bars)
```

```python
sector_sales = df.groupby('Occupation', as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc={'figure.figsize': (25,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sector_sales)
```

```python
product_sales = sns.countplot(x = 'Product_Category', data = df)

for bars in product_sales.containers:
    product_sales.bar_label(bars)
```

```python
sector_sales = df.groupby('Product_Category', as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize': (25,5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sector_sales)
```

```python
product_sales = sns.countplot(x = 'Product_ID', data = df)

for bars in product_sales.containers:
    product_sales.bar_label(bars)
```

```python
sector_sales = df.groupby('Product_ID', as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize': (25,5)})
sns.barplot(x = 'Product_ID', y = 'Orders', data = sector_sales)
```
